```python
import os
import math
import shutil
import random
from pathlib import Path

import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tqdm import tqdm
from sklearn.metrics import roc_curve, auc
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.applications.resnet50 import preprocess_input

# MTCNN is used as the face detection model.
# Why MTCNN?
#   - It detects faces better than simple Haar cascade.
#   - It is easy to use in Colab.
#   - It gives bounding boxes around faces.
try:
    from mtcnn import MTCNN
    MTCNN_AVAILABLE = True
except Exception:
    MTCNN_AVAILABLE = False
    print("MTCNN not available. Falling back to OpenCV Haar Cascade.")


# ============================================================
# 2. ENVIRONMENT SETUP
# ============================================================

print("Initializing environment...")

GPUS = tf.config.list_physical_devices("GPU")
if GPUS:
    print(f"GPU detected: {GPUS[0]}")
else:
    print("No GPU detected. Training will be slow on CPU.")

# AdaFace uses cosine, feature norms, margins, and scaling.
# float32 is safer than mixed precision for this type of angular-margin loss.
tf.keras.mixed_precision.set_global_policy("float32")

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)


# ============================================================
# 3. CONFIGURATION
# ============================================================

CONFIG = {
    # ---------------- Dataset download ----------------
    # These are Kaggle dataset slugs.
    "pins_kaggle_slug": "hereisburak/pins-face-recognition",
    "lfw_kaggle_slug": "jessicali9530/lfw-dataset",

    # Raw download directory.
    "raw_data_dir": "/content/raw_datasets",

    # Cleaned face-only dataset will be saved here.
    "processed_data_dir": "/content/processed_faces",

    # Test images should be placed here.
    # Put 10 external/new images inside this folder.
    "test_images_dir": "/content/test_10_images",

    # Face image size used by many face recognition models.
    "img_size": (112, 112),

    # Training details.
    "batch_size": 32,
    "epochs": 20, # Changed from 30 to 20
    "learning_rate": 1e-4,

    # Face embedding size.
    "embedding_dim": 512,

    # AdaFace hyperparameters.
    "adaface_scale": 64.0,
    "adaface_margin": 0.4,
    "adaface_h": 0.333,

    # Regularization / stability.
    "weight_decay": 1e-4,
    "dropout_rate": 0.4,
    "gradient_clip_norm": 5.0,
    "label_smoothing": 0.0,

    # Dataset split.
    "validation_split": 0.2,

    # Face detection settings.
    "detector_confidence": 0.90,
    "face_padding": 0.25,

    # Only keep classes with at least this many images.
    # This is useful for LFW because many identities have only 1 image.
    "min_images_per_class": 2,

    # Save model weights.
    "weights_path": "/content/best_resnet50_adaface.weights.h5",

    # Number of verification pairs for ROC.
    "roc_pairs": 2000,
}

IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}


# ============================================================
# 4. DATASET DOWNLOAD SECTION
# ============================================================


def download_kaggle_datasets():
    """
    Downloads the two datasets using kagglehub.

    Dataset 1:
        Pins Face Recognition Dataset
        Usually contains celebrity face folders.

    Dataset 2:
        LFW Dataset
        Labeled Faces in the Wild.

    Important:
        In Colab, kagglehub usually works without manually using kaggle.json.
        If download fails, use Kaggle API token setup:

            from google.colab import files
            files.upload()  # upload kaggle.json
            !mkdir -p ~/.kaggle
            !cp kaggle.json ~/.kaggle/
            !chmod 600 ~/.kaggle/kaggle.json
    """
    try:
        import kagglehub
    except Exception as e:
        raise ImportError("Install kagglehub first: !pip install kagglehub") from e

    raw_root = Path(CONFIG["raw_data_dir"])
    raw_root.mkdir(parents=True, exist_ok=True)

    pins_download_path = raw_root / "kaggle_pins"
    lfw_download_path = raw_root / "kaggle_lfw"

    # Check if Pins dataset is already downloaded
    if not pins_download_path.exists() or not any(pins_download_path.iterdir()):
        print("Downloading Pins Face Recognition dataset...")
        pins_kaggle_download_path = kagglehub.dataset_download(CONFIG["pins_kaggle_slug"], path=raw_root)
        # Rename or move to a consistent path if kagglehub creates a new one
        if pins_kaggle_download_path != pins_download_path:
            if pins_download_path.exists(): shutil.rmtree(pins_download_path)
            shutil.move(str(pins_kaggle_download_path), str(pins_download_path))
        print("Pins downloaded to:", pins_download_path)
    else:
        print("Pins dataset already exists. Skipping download.")

    # Check if LFW dataset is already downloaded
    if not lfw_download_path.exists() or not any(lfw_download_path.iterdir()):
        print("Downloading LFW dataset...")
        lfw_kaggle_download_path = kagglehub.dataset_download(CONFIG["lfw_kaggle_slug"], path=raw_root)
        # Rename or move to a consistent path if kagglehub creates a new one
        if lfw_download_path.exists(): shutil.rmtree(lfw_download_path)
        shutil.move(str(lfw_kaggle_download_path), str(lfw_download_path))
        print("LFW downloaded to:", lfw_download_path)
    else:
        print("LFW dataset already exists. Skipping download.")

    return Path(pins_download_path), Path(lfw_download_path)


# ============================================================
# 5. FIND DATASET ROOT FOLDERS
# ============================================================


def find_folder_with_identity_subfolders(root_path):
    """
    Kaggle datasets may unzip with different internal folder names.
    This function searches for a folder that contains many subfolders,
    where each subfolder represents one person.

    Example Pins path:
        .../105_classes_pins_dataset/pins_Aaron Paul/image.jpg

    Example LFW path:
        .../lfw-deepfunneled/George_W_Bush/image.jpg
    """
    root_path = Path(root_path)

    candidate_folders = []
    for folder in [root_path] + list(root_path.rglob("*")):
        if not folder.is_dir():
            continue

        subdirs = [x for x in folder.iterdir() if x.is_dir()]
        if len(subdirs) < 5:
            continue

        image_count_in_subdirs = 0
        for subdir in subdirs[:20]:
            image_count_in_subdirs += sum(
                1 for f in subdir.iterdir()
                if f.is_file() and f.suffix.lower() in IMAGE_EXTENSIONS
            )

        if image_count_in_subdirs > 0:
            candidate_folders.append((folder, len(subdirs), image_count_in_subdirs))

    if not candidate_folders:
        raise ValueError(f"Could not find identity-folder dataset inside {root_path}")

    # Pick largest face area.
    candidate_folders.sort(key=lambda b: b[2] * b[3], reverse=True)
    selected = candidate_folders[0][0]
    print(f"Selected dataset root: {selected}")
    return selected


# ============================================================
# 6. FACE DETECTION MODEL
# ============================================================

class FaceDetector:
    """
    Face detection wrapper.

    Primary detector:
        MTCNN

    Fallback detector:
        OpenCV Haar Cascade

    Why do face detection before training?
        The model should focus on the face, not the background.
        If one dataset has already-cropped faces and another has full images,
        the model may learn dataset-specific background patterns.

    Cropping faces makes both datasets more similar:
        full image -> face-only image -> 112x112 tensor
    """

    def __init__(self):
        if MTCNN_AVAILABLE:
            self.detector_name = "MTCNN"
            self.detector = MTCNN()
        else:
            self.detector_name = "OpenCV Haar"
            haar_path = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
            self.detector = cv2.CascadeClassifier(haar_path)

        print(f"Face detector used: {self.detector_name}")

    def detect_largest_face(self, image_rgb):
        """
        Returns bounding box of the largest detected face.

        Output format:
            x, y, w, h

        If no face is detected:
            returns None
        """
        if MTCNN_AVAILABLE:
            detections = self.detector.detect_faces(image_rgb)
            valid = []

            for d in detections:
                confidence = d.get("confidence", 0.0)
                if confidence < CONFIG["detector_confidence"]:
                    continue

                x, y, w, h = d["box"]
                if w <= 0 or h <= 0:
                    continue
                valid.append((x, y, w, h, confidence))

            if not valid:
                return None

            # Pick largest face area.
            valid.sort(key=lambda b: b[2] * b[3], reverse=True)
            x, y, w, h, _ = valid[0]
            return x, y, w, h

        else:
            gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
            boxes = self.detector.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
            if len(boxes) == 0:
                return None

            boxes = sorted(boxes, key=lambda b: b[2] * b[3], reverse=True)
            x, y, w, h = boxes[0]
            return x, y, w, h

    def crop_face(self, image_rgb, box):
        """
        Crops the face with padding.

        Why padding?
            A tight crop may remove forehead, chin, or ears.
            Small padding keeps useful identity information.
        """
        h_img, w_img, _ = image_rgb.shape
        x, y, w, h = box

        pad = CONFIG["face_padding"]
        pad_w = int(w * pad)
        pad_h = int(h * pad)

        x1 = max(0, x - pad_w)
        y1 = max(0, y - pad_h)
        x2 = min(w_img, x + w + pad_w)
        y2 = min(h_img, y + h + pad_h)

        face = image_rgb[y1:y2, x1:x2]
        return face

    def detect_crop_resize(self, image_path):
        """
        Reads an image, detects face, crops it, resizes to 112x112.

        If no face is detected:
            returns None
        """
        image_bgr = cv2.imread(str(image_path))
        if image_bgr is None:
            return None

        image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
        box = self.detect_largest_face(image_rgb)
        if box is None:
            return None

        face_rgb = self.crop_face(image_rgb, box)
        if face_rgb.size == 0:
            return None

        face_rgb = cv2.resize(face_rgb, CONFIG["img_size"])
        return face_rgb


# ============================================================
# 7. CREATE PROCESSED FACE DATASET
# ============================================================


def process_dataset_with_face_detection(dataset_root, output_root, prefix, detector):
    """
    Creates a cleaned dataset where every saved image is a detected face crop.

    Input:
        raw_dataset/person/image.jpg

    Output:
        processed_faces/prefix__person/image.jpg

    tqdm is used to show progress.
    """
    dataset_root = Path(dataset_root)
    output_root = Path(output_root)
    output_root.mkdir(parents=True, exist_ok=True)

    identity_dirs = [x for x in sorted(dataset_root.iterdir()) if x.is_dir()]

    total_images = []
    for identity_dir in identity_dirs:
        for image_path in identity_dir.rglob("*"):
            if image_path.suffix.lower() in IMAGE_EXTENSIONS:
                total_images.append((identity_dir.name, image_path))

    print(f"Processing {prefix}: {len(total_images)} images")

    saved_count = 0
    failed_count = 0

    for identity_name, image_path in tqdm(total_images, desc=f"Face detection: {prefix}"):
        face_rgb = detector.detect_crop_resize(image_path)

        if face_rgb is None:
            failed_count += 1
            continue

        clean_identity = f"{prefix}__{identity_name}"
        out_dir = output_root / clean_identity
        out_dir.mkdir(parents=True, exist_ok=True)

        out_path = out_dir / image_path.name
        face_bgr = cv2.cvtColor(face_rgb, cv2.COLOR_RGB2BGR)
        cv2.imwrite(str(out_path), face_bgr)
        saved_count += 1

    print(f"{prefix} saved faces: {saved_count}")
    print(f"{prefix} failed detections: {failed_count}")


# ============================================================
# 8. DATASET PREPARATION FUNCTION
# ============================================================


def prepare_datasets(force_reprocess=False):
    """
    Full dataset setup:
        1. Download datasets.
        2. Find actual identity-folder roots.
        3. Run face detection.
        4. Save cleaned face crops.

    Set force_reprocess=True if you want to delete old processed data
    and run face detection again.
    """
    processed_root = Path(CONFIG["processed_data_dir"])

    if force_reprocess and processed_root.exists():
        shutil.rmtree(processed_root)

    if processed_root.exists() and any(processed_root.iterdir()):
        print("Processed dataset already exists. Skipping face detection preprocessing.")
        return processed_root

    pins_download_path, lfw_download_path = download_kaggle_datasets()

    pins_root = find_folder_with_identity_subfolders(pins_download_path)
    lfw_root = find_folder_with_identity_subfolders(lfw_download_path)

    detector = FaceDetector()

    process_dataset_with_face_detection(
        dataset_root=pins_root,
        output_root=processed_root,
        prefix="pins",
        detector=detector,
    )

    process_dataset_with_face_detection(
        dataset_root=lfw_root,
        output_root=processed_root,
        prefix="lfw",
        detector=detector,
    )

    return processed_root


# ============================================================
# 9. SCAN PROCESSED DATASET
# ============================================================


def scan_processed_dataset(processed_root):
    """
    Reads processed face crops and returns:
        samples = [(image_path, identity_name), ...]
    """
    processed_root = Path(processed_root)
    samples = []

    for identity_dir in sorted(processed_root.iterdir()):
        if not identity_dir.is_dir():
            continue

        image_paths = [
            p for p in identity_dir.rglob("*")
            if p.suffix.lower() in IMAGE_EXTENSIONS
        ]

        # Remove identities with too few images.
        # LFW has many people with one image only. Those are not useful for verification.
        if len(image_paths) < CONFIG["min_images_per_class"]:
            continue

        for p in image_paths:
            samples.append((str(p), identity_dir.name))

    return samples


# ============================================================
# 10. TRAIN / VALIDATION SPLIT
# ============================================================


def split_per_identity(samples):
    """
    Splits each identity into train and validation.

    Why not random global split?
        Face recognition needs each identity represented properly.
        Per-identity splitting prevents some identities from disappearing from validation.
    """
    identity_to_images = {}
    for image_path, identity in samples:
        identity_to_images.setdefault(identity, []).append(image_path)

    train_items = []
    val_items = []

    for identity, paths in identity_to_images.items():
        random.shuffle(paths)
        label = class_to_index[identity]

        n_val = max(1, int(len(paths) * CONFIG["validation_split"]))
        val_paths = paths[:n_val]
        train_paths = paths[n_val:]

        train_items.extend([(p, label) for p in train_paths])
        val_items.extend([(p, label) for p in val_paths])

    random.shuffle(train_items)
    random.shuffle(val_items)
    return train_items, val_items


# ============================================================
# 11. TENSORFLOW PREPROCESSING
# ============================================================


def load_preprocessed_face(image_path, label, training=False):
    """
    Loads already face-cropped image and prepares it for ResNet-50.

    Preprocessing used:
        1. Read image.
        2. Decode as RGB.
        3. Resize to 112x112 again for safety.
        4. Training augmentation.
        5. ResNet-50 preprocess_input.

    Why preprocess_input?
        We use ImageNet pretrained ResNet-50 weights.
        ResNet-50 expects the same style of input normalization used during ImageNet training.
    """
    image = tf.io.read_file(image_path)
    image = tf.image.decode_image(image, channels=3, expand_animations=False)
    image = tf.image.resize(image, CONFIG["img_size"])
    image = tf.cast(image, tf.float32)

    if training:
        image = tf.image.random_flip_left_right(image)
        image = tf.image.random_brightness(image, max_delta=20.0)
        image = tf.image.random_contrast(image, lower=0.8, upper=1.2)
        image = tf.clip_by_value(image, 0.0, 255.0)

    image = preprocess_input(image)
    return image, label


# ============================================================
# 12. RESNET-50 + ADAFACE MODEL
# ============================================================

class AdaFaceLayer(layers.Layer):
    """
    AdaFace classification layer.

    ArcFace idea:
        Apply a fixed angular margin to the correct class.

    AdaFace idea:
        Apply an adaptive margin based on image quality.

    In simple words:
        Clear image  -> push harder
        Poor image   -> push carefully

    Feature norm is used as the quality indicator.
    """

    def __init__(self, num_classes, scale=64.0, margin=0.4, h=0.333, **kwargs):
        super().__init__(**kwargs)
        self.num_classes = num_classes
        self.scale = scale
        self.margin = margin
        self.h = h

        # Moving statistics for feature norm.
        self.moving_mean = tf.Variable(20.0, trainable=False, dtype=tf.float32)
        self.moving_std = tf.Variable(100.0, trainable=False, dtype=tf.float32)
        self.momentum = 0.99

    def build(self, input_shape):
        embedding_dim = input_shape[0][-1]
        self.W = self.add_weight(
            name="class_centers",
            shape=(embedding_dim, self.num_classes),
            initializer="glorot_uniform",
            trainable=True,
        )

    def call(self, inputs, labels, training=None):
        embeddings, feature_norms = inputs
        labels = tf.cast(labels, tf.int32)

        # Normalize class weights so dot product becomes cosine similarity.
        W_norm = tf.math.l2_normalize(self.W, axis=0)
        cosine = tf.matmul(embeddings, W_norm)
        cosine = tf.clip_by_value(cosine, -1.0 + 1e-7, 1.0 - 1e-7)

        raw_logits = cosine * self.scale

        # Feature norm as image-quality signal.
        safe_norms = tf.clip_by_value(feature_norms, 0.001, 100.0)

        if training:
            batch_mean = tf.reduce_mean(safe_norms)
            batch_std = tf.math.reduce_std(safe_norms)
            self.moving_mean.assign(self.momentum * self.moving_mean + (1.0 - self.momentum) * batch_mean)
            self.moving_std.assign(self.momentum * self.moving_std + (1.0 - self.momentum) * batch_std)

        normed_quality = (safe_norms - self.moving_mean) / (self.moving_std + 1e-6)

        # Stop gradient prevents the model from cheating by increasing norms artificially.
        normed_quality = tf.stop_gradient(normed_quality)
        normed_quality = tf.clip_by_value(normed_quality, -1.0, 1.0)

        adaptive_margin = self.margin * (1.0 + self.h * normed_quality)
        adaptive_margin = tf.reshape(adaptive_margin, [-1])

        one_hot = tf.one_hot(labels, self.num_classes, dtype=tf.float32)

        theta = tf.acos(cosine)
        theta_m = theta + tf.expand_dims(adaptive_margin, axis=1) * one_hot
        cosine_m = tf.cos(theta_m)

        penalized_logits = cosine_m * self.scale
        return penalized_logits, raw_logits


class ResNet50AdaFace(tf.keras.Model):
    """
    Training mode:
        model([images, labels]) -> AdaFace logits

    Inference mode:
        model(images) -> 512-dimensional normalized embeddings
    """

    def __init__(self, num_classes, embedding_dim=512):
        super().__init__()

        # ResNet-50 has 50 layers in its original architecture counting style.
        # We remove the original ImageNet classifier using include_top=False.
        self.resnet = tf.keras.applications.ResNet50(
            input_shape=(CONFIG["img_size"][0], CONFIG["img_size"][1], 3),
            include_top=False,
            weights="imagenet",
        )
        self.resnet.trainable = True

        self.gap = layers.GlobalAveragePooling2D()
        self.pre_bn = layers.BatchNormalization()
        self.dropout = layers.Dropout(CONFIG["dropout_rate"])
        self.embedding_dense = layers.Dense(
            embedding_dim,
            use_bias=False,
            kernel_initializer="he_normal",
            kernel_regularizer=regularizers.l2(CONFIG["weight_decay"]),
        )
        self.embedding_bn = layers.BatchNormalization()

        self.adaface = AdaFaceLayer(
            num_classes=num_classes,
            scale=CONFIG["adaface_scale"],
            margin=CONFIG["adaface_margin"],
            h=CONFIG["adaface_h"],
        )

    def extract_embeddings(self, images, training=None):
        x = self.resnet(images, training=training)
        x = self.gap(x)
        x = self.pre_bn(x, training=training)
        x = self.dropout(x, training=training)
        x = self.embedding_dense(x)
        x = self.embedding_bn(x, training=training)

        feature_norms = tf.norm(x, axis=1, keepdims=True)
        embeddings = tf.math.l2_normalize(x, axis=1)
        return embeddings, feature_norms

    def call(self, inputs, training=None):
        if isinstance(inputs, (list, tuple)) and len(inputs) == 2:
            images, labels = inputs
            embeddings, feature_norms = self.extract_embeddings(images, training=training)
            return self.adaface([embeddings, feature_norms], labels, training=training)

        images = inputs
        embeddings, _ = self.extract_embeddings(images, training=False)
        return embeddings


# ============================================================
# 13. PREPARE DATA
# ============================================================

processed_root = prepare_datasets(force_reprocess=False)
all_samples = scan_processed_dataset(processed_root)

if len(all_samples) == 0:
    raise ValueError("No processed face images found.")

identity_names = sorted(list({identity for _, identity in all_samples}))
class_to_index = {name: idx for idx, name in enumerate(identity_names)}
index_to_class = {idx: name for name, idx in class_to_index.items()}
num_classes = len(identity_names)

print(f"Total usable images: {len(all_samples)}")
print(f"Total usable identities/classes: {num_classes}")

train_items, val_items = split_per_identity(all_samples)

print(f"Train images: {len(train_items)}")
print(f"Validation images: {len(val_items)}")

train_paths = [p for p, _ in train_items]
train_labels = [y for _, y in train_items]
val_paths = [p for p, _ in val_items]
val_labels = [y for _, y in val_items]

train_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
train_ds = train_ds.shuffle(min(len(train_items), 10000), seed=SEED)
train_ds = train_ds.map(lambda p, y: load_preprocessed_face(p, y, training=True), num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.batch(CONFIG["batch_size"], drop_remainder=True)
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((val_paths, val_labels))
val_ds = val_ds.map(lambda p, y: load_preprocessed_face(p, y, training=False), num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.batch(CONFIG["batch_size"], drop_remainder=False)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)

steps_per_epoch = max(1, len(train_items) // CONFIG["batch_size"])
val_steps = max(1, math.ceil(len(val_items) / CONFIG["batch_size"]))


# ============================================================
# 14. LOSS, OPTIMIZER, METRICS
# ============================================================

model = ResNet50AdaFace(num_classes=num_classes, embedding_dim=CONFIG["embedding_dim"])

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(
    from_logits=True
)

optimizer = optimizers.Adam(learning_rate=CONFIG["learning_rate"])

train_loss_metric = tf.keras.metrics.Mean()
train_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy()
val_loss_metric = tf.keras.metrics.Mean()
val_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy()


def get_lr(epoch):
    # Simple step decay for 20 epochs
    if epoch < 10:
        return CONFIG["learning_rate"]
    else:
        return CONFIG["learning_rate"] * 0.1


@tf.function
def train_step(images, labels):
    with tf.GradientTape() as tape:
        penalized_logits, raw_logits = model([images, labels], training=True)
        loss = loss_fn(labels, penalized_logits)

        # Add L2 regularization losses.
        if model.losses:
            loss += tf.add_n(model.losses)

    grads = tape.gradient(loss, model.trainable_variables)

    # Stops gradient explosion.
    grads, global_norm = tf.clip_by_global_norm(grads, CONFIG["gradient_clip_norm"])

    optimizer.apply_gradients(zip(grads, model.trainable_variables))

    train_loss_metric.update_state(loss)
    train_acc_metric.update_state(labels, raw_logits)
    return global_norm


@tf.function
def val_step(images, labels):
    penalized_logits, raw_logits = model([images, labels], training=False)
    loss = loss_fn(labels, penalized_logits)

    if model.losses:
        loss += tf.add_n(model.losses)

    val_loss_metric.update_state(loss)
    val_acc_metric.update_state(labels, raw_logits)


# ============================================================
# 15. TRAINING WITH TQDM
# ============================================================

print("Starting training...")

best_val_acc = 0.0
epochs_no_improve = 0 # Counter for early stopping
patience = 5 # Number of epochs to wait before early stopping

for epoch in range(CONFIG["epochs"]):
    current_lr = get_lr(epoch)
    optimizer.learning_rate.assign(current_lr)

    for metric in [train_loss_metric, train_acc_metric, val_loss_metric, val_acc_metric]:
        metric.reset_state()

    print(f"\nEpoch {epoch + 1}/{CONFIG['epochs']} | LR: {current_lr:.8f}")

    grad_norms = []
    train_bar = tqdm(train_ds, total=steps_per_epoch, desc="Training", unit="batch")
    for images, labels in train_bar:
        grad_norm = train_step(images, labels)
        grad_norms.append(float(grad_norm.numpy()))

        train_bar.set_postfix({
            "loss": f"{train_loss_metric.result().numpy():.4f}",
            "acc": f"{train_acc_metric.result().numpy():.4f}",
            "grad": f"{np.mean(grad_norms):.2f}",
        })

    val_bar = tqdm(val_ds, total=val_steps, desc="Validation", unit="batch")
    for images, labels in val_bar:
        val_step(images, labels)
        val_bar.set_postfix({
            "val_loss": f"{val_loss_metric.result().numpy():.4f}",
            "val_acc": f"{val_acc_metric.result().numpy():.4f}",
        })

    val_acc = val_acc_metric.result().numpy()

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        model.save_weights(CONFIG["weights_path"])
        epochs_no_improve = 0 # Reset patience counter
        print(f"New best model saved: {CONFIG['weights_path']}")
    else:
        epochs_no_improve += 1
        print(f"Validation accuracy did not improve. Patience: {epochs_no_improve}/{patience}")

    print(
        f"Summary | "
        f"Train Loss: {train_loss_metric.result().numpy():.4f} | "
        f"Train Acc: {train_acc_metric.result().numpy() * 100:.2f}% | "
        f"Val Loss: {val_loss_metric.result().numpy():.4f} | "
        f"Val Acc: {val_acc * 100:.2f}% | "
        f"Mean Grad Norm: {np.mean(grad_norms):.2f}"
    )

    if epochs_no_improve >= patience:
        print(f"Early stopping triggered after {patience} epochs without improvement.")
        break # Stop training


# ============================================================
# 16. VALIDATION ROC EVALUATION
# ============================================================

print("\nExtracting validation embeddings...")

if os.path.exists(CONFIG["weights_path"]):
    model.load_weights(CONFIG["weights_path"])
    print("Loaded best saved weights.")

all_embeddings = []
all_labels = []

for images, labels in tqdm(val_ds, total=val_steps, desc="Validation embedding extraction"):
    embeddings = model(images, training=False).numpy()
    all_embeddings.append(embeddings)
    all_labels.extend(labels.numpy())

all_embeddings = np.vstack(all_embeddings)
all_labels = np.array(all_labels)


def build_random_pairs(labels, n_pairs=2000):
    label_to_indices = {}
    for idx, label in enumerate(labels):
        label_to_indices.setdefault(label, []).append(idx)

    positive_labels = [label for label, idxs in label_to_indices.items() if len(idxs) >= 2]
    unique_labels = list(label_to_indices.keys())

    pair_indices = []
    pair_targets = []

    half = n_pairs // 2

    for _ in range(half):
        label = random.choice(positive_labels)
        a, b = random.sample(label_to_indices[label], 2)
        pair_indices.append((a, b))
        pair_targets.append(1)

    for _ in range(half):
        label_a, label_b = random.sample(unique_labels, 2)
        a = random.choice(label_to_indices[label_a])
        b = random.choice(label_to_indices[label_b])
        pair_indices.append((a, b))
        pair_targets.append(0)

    return pair_indices, np.array(pair_targets)


n_pairs = min(CONFIG["roc_pairs"], max(100, len(all_embeddings)))
pair_indices, gt_labels = build_random_pairs(all_labels, n_pairs=n_pairs)

scores = []
for a, b in tqdm(pair_indices, desc="Computing ROC pair similarities"):
    sim = float(np.dot(all_embeddings[a], all_embeddings[b]))
    scores.append(sim)

scores = np.array(scores)
fpr, tpr, thresholds = roc_curve(gt_labels, scores)
roc_auc = auc(fpr, tpr)
best_idx = np.argmax(tpr - fpr)
best_threshold = thresholds[best_idx]

print(f"Validation ROC AUC: {roc_auc:.4f}")
print(f"Best threshold: {best_threshold:.4f}")

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, lw=2, label=f"ROC AUC = {roc_auc:.4f}")
plt.plot([0, 1], [0, 1], linestyle="--", label="Random baseline")
plt.scatter([fpr[best_idx]], [tpr[best_idx]], s=80, label="Best threshold")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Validation Face Verification ROC")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# ============================================================
# 17. BUILD CLASS PROTOTYPES FOR TESTING
# ============================================================

print("\nBuilding class prototypes from training images...")

# A prototype is the average embedding of one identity.
# During testing, we compare a new image embedding with all prototypes.
# The nearest prototype becomes the predicted identity.
class_embedding_sum = {}
class_embedding_count = {}

prototype_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
prototype_ds = prototype_ds.map(lambda p, y: load_preprocessed_face(p, y, training=False), num_parallel_calls=tf.data.AUTOTUNE)
prototype_ds = prototype_ds.batch(CONFIG["batch_size"])
prototype_ds = prototype_ds.prefetch(tf.data.AUTOTUNE)

for images, labels in tqdm(prototype_ds, desc="Prototype embedding extraction"):
    embeddings = model(images, training=False).numpy()
    labels_np = labels.numpy()

    for emb, label in zip(embeddings, labels_np):
        if label not in class_embedding_sum:
            class_embedding_sum[label] = np.zeros_like(emb)
            class_embedding_count[label] = 0

        class_embedding_sum[label] += emb
        class_embedding_count[label] += 1

class_prototypes = {}
for label, emb_sum in class_embedding_sum.items():
    proto = emb_sum / class_embedding_count[label]
    proto = proto / (np.linalg.norm(proto) + 1e-8)
    class_prototypes[label] = proto

print(f"Built prototypes for {len(class_prototypes)} classes.")


# ============================================================
# 18. TESTING WITH 10 NEW IMAGES + FACE DETECTION + DISPLAY
# ============================================================


def preprocess_external_test_image(image_path, detector):
    """
    This is the testing-time face detection pipeline.

    For every test image:
        1. Detect face.
        2. Crop face.
        3. Resize to 112x112.
        4. Apply ResNet preprocessing.
        5. Send to model.
    """
    face_rgb = detector.detect_crop_resize(image_path)
    if face_rgb is None:
        return None, None

    image = tf.cast(face_rgb, tf.float32)
    image = preprocess_input(image)
    image = tf.expand_dims(image, axis=0)
    return image, face_rgb


def predict_identity_from_image(image_path, detector, threshold=None):
    """
    Predicts identity for one external test image.

    If threshold is provided:
        similarity below threshold -> Unknown
    """
    image_tensor, face_rgb = preprocess_external_test_image(image_path, detector)

    if image_tensor is None:
        return {
            "image_path": image_path,
            "face_rgb": None,
            "predicted_name": "No face detected",
            "similarity": None,
        }

    emb = model(image_tensor, training=False).numpy()[0]
    emb = emb / (np.linalg.norm(emb) + 1e-8)

    best_label = None
    best_score = -1.0

    for label, proto in class_prototypes.items():
        score = float(np.dot(emb, proto))
        if score > best_score:
            best_score = score
            best_label = label

    predicted_name = index_to_class[best_label]

    if threshold is not None and best_score < threshold:
        predicted_name = "Unknown"

    return {
        "image_path": image_path,
        "face_rgb": face_rgb,
        "predicted_name": predicted_name,
        "similarity": best_score,
    }


def test_and_display_10_images():
    """
    Tests 10 new images from CONFIG['test_images_dir'].

    Put images here:
        /content/test_10_images/

    Example:
        /content/test_10_images/test1.jpg
        /content/test_10_images/test2.jpg
        ...
    """
    test_dir = Path(CONFIG["test_images_dir"])
    test_dir.mkdir(parents=True, exist_ok=True)

    test_images = [
        p for p in sorted(test_dir.iterdir())
        if p.suffix.lower() in IMAGE_EXTENSIONS
    ]

    if len(test_images) == 0:
        print(f"No test images found in {test_dir}")
        print("Upload 10 new images into this folder and run test_and_display_10_images() again.")
        return

    test_images = test_images[:10]
    detector = FaceDetector()

    results = []
    for image_path in tqdm(test_images, desc="Testing 10 new images"):
        result = predict_identity_from_image(
            image_path=image_path,
            detector=detector,
            threshold=best_threshold,
        )
        results.append(result)

    cols = 5
    rows = math.ceil(len(results) / cols)
    plt.figure(figsize=(18, 4 * rows))

    for i, result in enumerate(results):
        plt.subplot(rows, cols, i + 1)

        if result["face_rgb"] is None:
            original_bgr = cv2.imread(str(result["image_path"]))
            original_rgb = cv2.cvtColor(original_bgr, cv2.COLOR_BGR2RGB)
            plt.imshow(original_rgb)
            title = "No face detected"
        else:
            plt.imshow(result["face_rgb"])
            sim_text = f"{result['similarity']:.3f}" if result["similarity"] is not None else "N/A"
            title = f"{result['predicted_name']}\nSim: {sim_text}"

        plt.title(title, fontsize=9)
        plt.axis("off")

    plt.tight_layout()
    plt.show()

    print("\nTest Results:")
    for r in results:
        print(f"{Path(r['image_path']).name} -> {r['predicted_name']} | similarity={r['similarity']}")


# Run final 10-image testing.
test_and_display_10_images()


# ============================================================
# 19. IMPORTANT NOTES
# ============================================================
# 1. If the model predicts 'Unknown' for many images:
#       - The person may not exist in training classes.
#       - The threshold may be too strict.
#       - Face crop may be bad.
#
# 2. If validation accuracy is low:
#       - Use more images per identity.
#       - Train longer.
#       - Freeze ResNet early layers for first few epochs, then unfreeze.
#       - Use better aligned face crops.
#
# 3. If face detection is slow:
#       - MTCNN is accurate but slower.
#       - Preprocessing is done once and saved.
#       - After processed_faces exists, the code skips detection unless force_reprocess=True.
#
# 4. Why testing uses prototypes instead of Dense classifier weights:
#       - Face recognition is embedding-based.
#       - We compare new face embedding with known identity embeddings.
#       - This is closer to real face recognition than normal softmax classification.
#
# 5. Real-world face recognition usually needs:
#       - Face detection
#       - Face alignment using eye landmarks
#       - Embedding extraction
#       - Similarity search
#       - Threshold-based unknown detection
# ============================================================
```

```python
import os
import math
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from pathlib import Path
from tqdm import tqdm
from tensorflow.keras.applications.resnet50 import preprocess_input

# Folder where you uploaded your test images
TEST_PAIR_DIR = Path("/content/test")

IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}


def preprocess_test_image_for_embedding(image_path, detector):
    """
    Takes one image path:
        image -> face detection -> face crop -> resize -> ResNet preprocess -> tensor

    Returns:
        image_tensor: model-ready tensor of shape (1, 112, 112, 3)
        face_rgb: cropped face image for display
    """

    # Use your existing FaceDetector class
    face_rgb = detector.detect_crop_resize(image_path)

    if face_rgb is None:
        return None, None

    # Convert face crop to float32
    image = tf.cast(face_rgb, tf.float32)

    # Apply same preprocessing used during training
    image = preprocess_input(image)

    # Add batch dimension
    image = tf.expand_dims(image, axis=0)

    return image, face_rgb


def extract_embeddings_from_test_folder(test_dir):
    """
    Loops through all images in test folder and creates embeddings.
    """

    detector = FaceDetector()

    image_paths = [
        p for p in sorted(test_dir.iterdir())
        if p.suffix.lower() in IMAGE_EXTENSIONS
    ]

    print("Total test images found:", len(image_paths))

    if len(image_paths) < 2:
        raise ValueError("Need at least 2 images for pair matching.")

    embeddings = []
    face_images = []
    valid_paths = []

    for image_path in tqdm(image_paths, desc="Extracting test embeddings"):
        image_tensor, face_rgb = preprocess_test_image_for_embedding(image_path, detector)

        if image_tensor is None:
            print("No face detected:", image_path.name)
            continue

        # Generate 512-dimensional embedding
        emb = model(image_tensor, training=False).numpy()[0]

        # L2 normalize again for safety
        emb = emb / (np.linalg.norm(emb) + 1e-8)

        embeddings.append(emb)
        face_images.append(face_rgb)
        valid_paths.append(image_path)

    embeddings = np.array(embeddings)

    print("Valid face images:", len(valid_paths))

    return valid_paths, face_images, embeddings


def find_best_pair_for_each_image(embeddings):
    """
    Computes cosine similarity between every image and every other image.

    Since embeddings are L2 normalized:
        cosine_similarity = dot product
    """

    # Similarity matrix: shape = (N, N)
    similarity_matrix = np.dot(embeddings, embeddings.T)

    # Ignore self similarity by setting diagonal to -1
    np.fill_diagonal(similarity_matrix, -1.0)

    best_pairs = []

    for i in range(len(embeddings)):
        best_j = np.argmax(similarity_matrix[i])
        best_score = similarity_matrix[i][best_j]

        best_pairs.append((i, best_j, best_score))

    return best_pairs, similarity_matrix


def display_best_pairs(valid_paths, face_images, best_pairs):
    """
    Displays each image with its best matching pair.
    """

    n = len(best_pairs)

    plt.figure(figsize=(10, 4 * n))

    for row, (i, j, score) in enumerate(best_pairs):
        # Left image
        plt.subplot(n, 2, row * 2 + 1)
        plt.imshow(face_images[i])
        plt.title(f"Query\n{valid_paths[i].name}", fontsize=10)
        plt.axis("off")

        # Right best match
        plt.subplot(n, 2, row * 2 + 2)
        plt.imshow(face_images[j])
        plt.title(
            f"Best Match\n{valid_paths[j].name}\nCosine Sim: {score:.4f}",
            fontsize=10
        )
        plt.axis("off")

    plt.tight_layout()
    plt.show()


def run_test_folder_pair_matching():
    """
    Main function.
    """

    valid_paths, face_images, embeddings = extract_embeddings_from_test_folder(TEST_PAIR_DIR)

    if len(valid_paths) < 2:
        raise ValueError("Less than 2 valid face images found.")

    best_pairs, similarity_matrix = find_best_pair_for_each_image(embeddings)

    print("\nBest match for each image:")
    for i, j, score in best_pairs:
        print(f"{valid_paths[i].name}  -->  {valid_paths[j].name}  | similarity = {score:.4f}")

    display_best_pairs(valid_paths, face_images, best_pairs)

    return valid_paths, face_images, embeddings, best_pairs, similarity_matrix


# Run it
valid_paths, face_images, embeddings, best_pairs, similarity_matrix = run_test_folder_pair_matching()
```
