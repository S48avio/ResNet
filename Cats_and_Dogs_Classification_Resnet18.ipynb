```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║         ResNet-18 — Cat vs Dog Binary Classification                ║
# ║         Adapted for: bhavikjikadara/dog-and-cat-classification      ║
# ╚══════════════════════════════════════════════════════════════════════╝

import os
import zipfile
import shutil
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tqdm import tqdm

print("=" * 55)
print("  ResNet-18 — Cat vs Dog Classification")
print("=" * 55)
print(f"TensorFlow : {tf.__version__}")
print(f"GPU        : {tf.config.list_physical_devices('GPU')}")

# ══════════════════════════════════════════════════════
# STEP 1: DOWNLOAD AND SETUP
# ══════════════════════════════════════════════════════

os.environ['KAGGLE_USERNAME'] = 'saviosunny4890123567'
os.environ['KAGGLE_KEY']      = 'KGAT_3ecb972a45bc934fa04b198cbc9ad0c1'

print("\nDownloading dataset...")
os.system('kaggle datasets download -d bhavikjikadara/dog-and-cat-classification-dataset -q')

print("Extracting...")
with zipfile.ZipFile('dog-and-cat-classification-dataset.zip', 'r') as z:
    z.extractall('pet_data')

# Your dataset's actual folder structure
# Kaggle gives: training_set/training_set/cats and dogs
#               test_set/test_set/cats and dogs
# Corrected to reflect typical extraction behavior (no double nested folder)
TRAIN_DIR = 'pet_data/PetImages'
TEST_DIR  = 'pet_data/PetImages'

# Verify structure exists
for folder in [TRAIN_DIR, TEST_DIR]:
    if os.path.exists(folder):
        classes = os.listdir(folder)
        print(f"  ✅ Found: {folder}  →  classes: {classes}")
    else:
        print(f"  ❌ Not found: {folder}")

# ══════════════════════════════════════════════════════
# STEP 2: CONFIGURATION
# ══════════════════════════════════════════════════════

CONFIG = {
    'img_size'    : (224, 224),   # ResNet standard input size
    'batch_size'  : 32,
    'epochs'      : 20,
    'lr'          : 0.001,
    'val_split'   : 0.15,         # 15% of training set → validation
    'test_count'  : 10,           # 10 images for visual test display
    'seed'        : 42,

    # Paths
    'train_dir'   : TRAIN_DIR,
    'test_dir'    : TEST_DIR,
    'val_dir'     : '/content/val_set',         # we create this
    'visual_dir'  : '/content/visual_test',     # 10 images for display
    'weights_path': '/content/best_resnet18.keras',
}

# ══════════════════════════════════════════════════════
# STEP 3: CREATE VALIDATION + VISUAL TEST SPLIT
#
# Your dataset already has train and test folders.
# We need to:
#   1. Carve out 15% of training data → validation folder
#   2. Pick 10 images from test folder → visual display
#
# WHY SEPARATE VALIDATION?
#   Training loss tells you how well the model fits training data.
#   Validation loss tells you if it GENERALIZES to new images.
#   Without validation you can't detect overfitting early.
# ══════════════════════════════════════════════════════

def create_val_split(train_dir, val_dir, val_split=0.15, seed=42):
    """
    Move val_split fraction of training images to validation folder.
    Does not touch the original test set.
    """
    random.seed(seed)

    # Your dataset uses 'cats' and 'dogs' (plural)
    classes = [c for c in os.listdir(train_dir)
               if os.path.isdir(os.path.join(train_dir, c))]
    print(f"\nClasses found: {classes}")

    for cls in classes:
        src_cls = os.path.join(train_dir, cls)
        dst_cls = os.path.join(val_dir,   cls)
        os.makedirs(dst_cls, exist_ok=True)

        images = [f for f in os.listdir(src_cls)
                  if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        random.shuffle(images)

        n_val = int(len(images) * val_split)
        val_images = images[:n_val]

        for img in tqdm(val_images, desc=f"  Val ← {cls}", leave=False):
            shutil.move(
                os.path.join(src_cls, img),
                os.path.join(dst_cls, img),
            )

        remaining = len(images) - n_val
        print(f"  {cls}: {remaining} train | {n_val} val")

    return classes


def prepare_visual_test(test_dir, visual_dir, n=10, seed=42):
    """
    Copy n images from the test set into a separate folder for display.
    Picks n//2 cats and n//2 dogs.
    """
    random.seed(seed)
    classes = [c for c in os.listdir(test_dir)
               if os.path.isdir(os.path.join(test_dir, c))]

    for cls in classes:
        src = os.path.join(test_dir, cls)
        dst = os.path.join(visual_dir, cls)
        os.makedirs(dst, exist_ok=True)

        images = [f for f in os.listdir(src)
                  if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        random.shuffle(images)
        chosen = images[:n // len(classes)]

        for img in chosen:
            shutil.copy(os.path.join(src, img), os.path.join(dst, img))

    total = sum(
        len(os.listdir(os.path.join(visual_dir, c)))
        for c in os.listdir(visual_dir)
        if os.path.isdir(os.path.join(visual_dir, c))
    )
    print(f"  Visual test: {total} images copied to {visual_dir}")


# Run only if not already done
if not os.path.exists(CONFIG['val_dir']):
    print("\nCreating validation split from training data...")
    classes = create_val_split(
        train_dir=CONFIG['train_dir'],
        val_dir=CONFIG['val_dir'],
        val_split=CONFIG['val_split'],
        seed=CONFIG['seed'],
    )
else:
    print("Validation split already exists. Skipping.")

if not os.path.exists(CONFIG['visual_dir']):
    print("\nPreparing visual test images...")
    prepare_visual_test(
        test_dir=CONFIG['test_dir'],
        visual_dir=CONFIG['visual_dir'],
        n=CONFIG['test_count'],
        seed=CONFIG['seed'],
    )
else:
    print("Visual test images already prepared. Skipping.")

# ══════════════════════════════════════════════════════
# STEP 4: DATA PIPELINE
#
# class_mode='binary' → labels are 0.0 or 1.0
# Your folder names 'cats' and 'dogs' → Keras assigns:
#   cats = 0  (alphabetical order)
#   dogs = 1
# Sigmoid output > 0.5 → dog, < 0.5 → cat
# ══════════════════════════════════════════════════════

print("\nBuilding data pipeline...")

train_datagen = ImageDataGenerator(
    rescale=1./255,              # pixel values 0-255 → 0.0-1.0
    rotation_range=15,           # random rotation ±15°
    width_shift_range=0.1,       # random horizontal shift
    height_shift_range=0.1,      # random vertical shift
    horizontal_flip=True,        # mirror image (cats/dogs are symmetric)
    brightness_range=[0.8, 1.2], # simulate different lighting
    zoom_range=0.1,              # slight zoom variation
    fill_mode='nearest',         # fill empty areas after transforms
)

# No augmentation for val/test — only normalize
eval_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    CONFIG['train_dir'],
    target_size=CONFIG['img_size'],
    batch_size=CONFIG['batch_size'],
    class_mode='binary',   # 0 = cats, 1 = dogs
    shuffle=True,
    seed=CONFIG['seed'],
)

val_gen = eval_datagen.flow_from_directory(
    CONFIG['val_dir'],
    target_size=CONFIG['img_size'],
    batch_size=CONFIG['batch_size'],
    class_mode='binary',
    shuffle=False,
)

# Full test set evaluation (all test images)
test_gen = eval_datagen.flow_from_directory(
    CONFIG['test_dir'],
    target_size=CONFIG['img_size'],
    batch_size=CONFIG['batch_size'],
    class_mode='binary',
    shuffle=False,
)

# Store class mapping: {0: 'cats', 1: 'dogs'}
class_indices = train_gen.class_indices           # {'cats': 0, 'dogs': 1}
idx_to_class  = {v: k for k, v in class_indices.items()}

print(f"\nClass mapping : {class_indices}")
print(f"Train images  : {train_gen.samples}")
print(f"Val images    : {val_gen.samples}")
print(f"Test images   : {test_gen.samples}")
print(f"Steps/epoch   : {train_gen.samples // CONFIG['batch_size']}")

# ══════════════════════════════════════════════════════
# STEP 5: RESNET-18 ARCHITECTURE (built from scratch)
#
# ResNet-18 layer count:
#   Conv1          : 1
#   Stage 1 (×2)   : 2 blocks × 2 conv = 4
#   Stage 2 (×2)   : 2 blocks × 2 conv = 4
#   Stage 3 (×2)   : 2 blocks × 2 conv = 4
#   Stage 4 (×2)   : 2 blocks × 2 conv = 4
#   FC             : 1
#   Total          : 1+4+4+4+4+1 = 18 ✅
#
# BASIC BLOCK (not bottleneck):
#   Conv→BN→ReLU→Conv→BN→Add(skip)→ReLU
#   Skip connection = gradient highway = no vanishing gradients
#
# WHY SIGMOID OUTPUT?
#   1 neuron → sigmoid → P(dog) ∈ (0,1)
#   > 0.5 = dog, < 0.5 = cat
#   P(cat) = 1 - P(dog)
# ══════════════════════════════════════════════════════

def basic_block(x, filters, stride=1):
    """
    One basic residual block.
    Two 3×3 convolutions with a skip connection.

    Input x  ──► Conv3×3 ──► BN ──► ReLU ──► Conv3×3 ──► BN ──► Add ──► ReLU
              │                                                      ↑
              └──────────────── skip (identity or 1×1 proj) ────────┘
    """
    shortcut = x

    # First conv: stride controls downsampling
    x = layers.Conv2D(filters, 3, strides=stride,
                      padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)        # hidden ReLU inside block

    # Second conv: always stride=1
    x = layers.Conv2D(filters, 3, strides=1,
                      padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    # NO ReLU here — apply after adding skip

    # Projection shortcut if shapes differ
    if stride != 1 or shortcut.shape[-1] != filters:
        shortcut = layers.Conv2D(filters, 1, strides=stride,
                                 padding='same', use_bias=False)(shortcut)
        shortcut = layers.BatchNormalization()(shortcut)

    # Add skip connection then activate
    x = layers.Add()([x, shortcut])
    x = layers.ReLU()(x)        # output ReLU on combined signal

    return x


def build_resnet18(input_shape=(224, 224, 3)):
    """
    ResNet-18 for binary classification.

    Spatial size journey:
      224×224 → Conv1+Pool → 56×56
      56×56   → Stage 1    → 56×56  (64 channels)
      56×56   → Stage 2    → 28×28  (128 channels)
      28×28   → Stage 3    → 14×14  (256 channels)
      14×14   → Stage 4    → 7×7    (512 channels)
      7×7     → GAP        → 512
      512     → Dense(256) → 256
      256     → Dense(1)   → 1  ← sigmoid output = P(dog)
    """
    inputs = layers.Input(shape=input_shape, name='input')

    # ── Conv1: 7×7, stride 2 → 112×112 ────────────────────────────────
    x = layers.Conv2D(64, 7, strides=2, padding='same',
                      use_bias=False, name='conv1')(inputs)
    x = layers.BatchNormalization(name='bn1')(x)
    x = layers.ReLU(name='relu1')(x)

    # MaxPool: 112×112 → 56×56
    x = layers.MaxPooling2D(3, strides=2, padding='same')(x)

    # ── Stage 1: 2 blocks, 64 filters, 56×56 ───────────────────────────
    # Learns: edges, fur texture, basic patterns
    x = basic_block(x, 64,  stride=1)
    x = basic_block(x, 64,  stride=1)

    # ── Stage 2: 2 blocks, 128 filters, 28×28 ──────────────────────────
    # Learns: ear shapes, snout/whisker patterns
    x = basic_block(x, 128, stride=2)   # stride=2 → 56→28
    x = basic_block(x, 128, stride=1)

    # ── Stage 3: 2 blocks, 256 filters, 14×14 ──────────────────────────
    # Learns: face structure, body shape
    x = basic_block(x, 256, stride=2)   # stride=2 → 28→14
    x = basic_block(x, 256, stride=1)

    # ── Stage 4: 2 blocks, 512 filters, 7×7 ────────────────────────────
    # Learns: high-level cat vs dog features
    x = basic_block(x, 512, stride=2)   # stride=2 → 14→7
    x = basic_block(x, 512, stride=1)

    # ── Global Average Pool: 7×7×512 → 512 ────────────────────────────
    x = layers.GlobalAveragePooling2D(name='gap')(x)

    # ── Dense head ─────────────────────────────────────────────────────
    x = layers.Dense(256, activation='relu', name='fc1')(x)
    x = layers.Dropout(0.5)(x)   # drop 50% → prevents overfitting

    # ── Sigmoid output: P(dog) ─────────────────────────────────────────
    # 1 neuron, sigmoid → value in (0, 1)
    # threshold 0.5: above = dog, below = cat
    output = layers.Dense(1, activation='sigmoid', name='output')(x)

    return models.Model(inputs, output, name='ResNet18')


# Build
model = build_resnet18(input_shape=(*CONFIG['img_size'], 3))
model.summary()
print(f"\nTotal parameters: {model.count_params():,}")

# ══════════════════════════════════════════════════════
# STEP 6: COMPILE
#
# binary_crossentropy: correct loss for sigmoid output
# Adam: fast, adaptive — good for classification
# ══════════════════════════════════════════════════════

model.compile(
    optimizer=optimizers.Adam(learning_rate=CONFIG['lr']),
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall'),
    ],
)
print("\nModel compiled.")
print("  Loss    : binary_crossentropy")
print("  Output  : sigmoid → P(dog)")
print("  Metrics : accuracy, AUC, precision, recall")

# ══════════════════════════════════════════════════════
# STEP 7: TQDM PROGRESS CALLBACK
# ══════════════════════════════════════════════════════

class TqdmCallback(tf.keras.callbacks.Callback):
    """Shows per-batch progress bar with live loss and accuracy."""

    def on_epoch_begin(self, epoch, logs=None):
        self.pbar = tqdm(
            total=self.params['steps'],
            desc=f"Epoch {epoch+1:02d}/{CONFIG['epochs']}",
            unit="batch",
            bar_format="{l_bar}{bar:40}{r_bar}",
            colour='blue',
        )

    def on_train_batch_end(self, batch, logs=None):
        self.pbar.set_postfix({
            'loss': f"{logs.get('loss', 0):.4f}",
            'acc' : f"{logs.get('accuracy', 0)*100:.1f}%",
        })
        self.pbar.update(1)

    def on_epoch_end(self, epoch, logs=None):
        self.pbar.close()
        val_acc = logs.get('val_accuracy', 0) * 100
        val_auc = logs.get('val_auc', 0)
        print(
            f"  → val_acc: {val_acc:.1f}%  "
            f"val_auc: {val_auc:.4f}  "
            f"lr: {float(self.model.optimizer.learning_rate):.6f}"
        )

# ══════════════════════════════════════════════════════
# STEP 8: TRAIN
# ══════════════════════════════════════════════════════

callbacks = [
    TqdmCallback(),

    tf.keras.callbacks.ModelCheckpoint(
        CONFIG['weights_path'],
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=0,
    ),

    tf.keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=5,
        restore_best_weights=True,
        verbose=1,
    ),

    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-6,
        verbose=1,
    ),
]

print("\n" + "="*55)
print("TRAINING")
print("="*55)

history = model.fit(
    train_gen,
    epochs=CONFIG['epochs'],
    validation_data=val_gen,
    callbacks=callbacks,
    verbose=0,
)

best_acc = max(history.history['val_accuracy'])
print(f"\n✅ Training done. Best val accuracy: {best_acc*100:.2f}%")

# ══════════════════════════════════════════════════════
# STEP 9: TRAINING CURVES
# ══════════════════════════════════════════════════════

epochs_ran = range(1, len(history.history['loss']) + 1)
fig, axes  = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('ResNet-18 Training — Cat vs Dog', fontsize=13, fontweight='bold')

axes[0].plot(epochs_ran, history.history['loss'],         label='Train', color='#185FA5', lw=2)
axes[0].plot(epochs_ran, history.history['val_loss'],     label='Val',   color='#E24B4A', lw=2)
axes[0].set_title('Binary Cross-Entropy Loss')
axes[0].set_xlabel('Epoch'); axes[0].legend(); axes[0].grid(True, alpha=0.3)

axes[1].plot(epochs_ran, [a*100 for a in history.history['accuracy']],     label='Train', color='#185FA5', lw=2)
axes[1].plot(epochs_ran, [a*100 for a in history.history['val_accuracy']], label='Val',   color='#E24B4A', lw=2)
axes[1].set_title('Accuracy (%)'); axes[1].set_xlabel('Epoch')
axes[1].legend(); axes[1].grid(True, alpha=0.3)

axes[2].plot(epochs_ran, history.history['auc'],     label='Train AUC', color='#185FA5', lw=2)
axes[2].plot(epochs_ran, history.history['val_auc'], label='Val AUC',   color='#E24B4A', lw=2)
axes[2].set_title('AUC (1.0 = perfect)')
axes[2].set_xlabel('Epoch'); axes[2].legend(); axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/content/training_curves.png', dpi=150, bbox_inches='tight')
plt.show()

# ══════════════════════════════════════════════════════
# STEP 10: FULL TEST SET EVALUATION
# ══════════════════════════════════════════════════════

print("\n" + "="*55)
print("FULL TEST SET EVALUATION")
print("="*55)

results = model.evaluate(test_gen, verbose=0)
metric_names = ['loss', 'accuracy', 'auc', 'precision', 'recall']
for name, val in zip(metric_names, results):
    print(f"  {name:<12}: {val:.4f}" if name == 'loss'
          else f"  {name:<12}: {val*100:.2f}%")

# ══════════════════════════════════════════════════════
# STEP 11: VISUAL TEST — 10 IMAGES WITH DISPLAY
#
# Loads each of the 10 visual test images.
# Runs prediction.
# Shows image grid with:
#   - True label (from folder name)
#   - Predicted label
#   - Confidence %
#   - Green border = correct, Red border = wrong
# ══════════════════════════════════════════════════════

print("\n" + "="*55)
print("VISUAL TEST — 10 Images")
print("="*55)

# Emoji map for display
EMOJI = {'cats': '🐱 Cat', 'dogs': '🐶 Dog'}

def predict_image(img_path):
    """
    Load, preprocess, and predict one image.
    """
    img   = tf.keras.utils.load_img(img_path, target_size=CONFIG['img_size'])
    arr   = tf.keras.utils.img_to_array(img) / 255.0
    batch = np.expand_dims(arr, 0)

    prob  = float(model.predict(batch, verbose=0)[0][0])
    # prob = P(dog) from sigmoid
    # dogs = class 1, cats = class 0

    if prob >= 0.5:
        pred_class = 'dogs'
        confidence = prob
    else:
        pred_class = 'cats'
        confidence = 1.0 - prob

    return pred_class, confidence, arr   # arr is normalized for display

# Collect test images
visual_images = []
for cls in sorted(os.listdir(CONFIG['visual_dir'])):
    cls_dir = os.path.join(CONFIG['visual_dir'], cls)
    if not os.path.isdir(cls_dir):
        continue
    for img_file in sorted(os.listdir(cls_dir)):
        if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            visual_images.append((
                os.path.join(cls_dir, img_file),
                cls   # true label = folder name
            ))

visual_images = visual_images[:CONFIG['test_count']]

# Run predictions
print(f"Predicting on {len(visual_images)} images...")
preds = []
for img_path, true_cls in tqdm(visual_images, desc="Predicting", unit="img"):
    pred_cls, conf, img_arr = predict_image(img_path)
    is_correct = (pred_cls == true_cls)
    preds.append({
        'img'    : img_arr,
        'true'   : true_cls,
        'pred'   : pred_cls,
        'conf'   : conf,
        'correct': is_correct,
    })

# ── Draw grid ─────────────────────────────────────────
n_cols = 5
n_rows = (len(preds) + n_cols - 1) // n_cols

fig, axes = plt.subplots(
    n_rows, n_cols,
    figsize=(n_cols * 4, n_rows * 4.5)
)
fig.suptitle(
    'ResNet-18 — Cat vs Dog  |  10 Test Images\n'
    'Green border = Correct Prediction   |   Red border = Wrong Prediction',
    fontsize=13, fontweight='bold', y=1.01
)

correct_total = 0

for idx in range(n_rows * n_cols):
    row = idx // n_cols
    col = idx  % n_cols
    ax  = axes[row][col] if n_rows > 1 else axes[col]
    ax.axis('off')

    if idx >= len(preds):
        continue

    r = preds[idx]
    ax.imshow(r['img'])

    # Colored border
    border = '#2ECC71' if r['correct'] else '#E74C3C'
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_edgecolor(border)
        spine.set_linewidth(6)

    # Title text
    color = '#1A8A35' if r['correct'] else '#C0392B'
    status = '✅' if r['correct'] else '❌'
    ax.set_title(
        f"{status}  True : {EMOJI[r['true']]}\n"
        f"     Pred : {EMOJI[r['pred']]}\n"
        f"     Conf : {r['conf']*100:.1f}%",
        fontsize=9.5,
        color=color,
        fontweight='bold',
        pad=6,
    )

    if r['correct']:
        correct_total += 1

# Legend
cat_patch = mpatches.Patch(color='#185FA5', label='🐱 Cat')
dog_patch = mpatches.Patch(color='#E24B4A', label='🐶 Dog')
ok_patch  = mpatches.Patch(color='#2ECC71', label='✅ Correct')
err_patch = mpatches.Patch(color='#E74C3C', label='❌ Wrong')
fig.legend(
    handles=[cat_patch, dog_patch, ok_patch, err_patch],
    loc='lower center', ncol=4, fontsize=11,
    bbox_to_anchor=(0.5, -0.03), frameon=True,
)

plt.tight_layout()
plt.savefig('/content/test_results.png', dpi=150, bbox_inches='tight')
plt.show()

# ── Print summary table ────────────────────────────────
print(f"\n{'='*55}")
print(f"VISUAL TEST RESULTS")
print(f"{'='*55}")
print(f"{'#':<4} {'True':<8} {'Predicted':<10} {'Confidence':<12} {'Result'}")
print("-" * 55)
for i, r in enumerate(preds):
    status = "✅ Correct" if r['correct'] else "❌ Wrong"
    print(f"{i+1:<4} {r['true']:<8} {r['pred']:<10} "
          f"{r['conf']*100:>6.1f}%      {status}")

print(f"\n  Score   : {correct_total} / {len(preds)} correct")
print(f"  Accuracy: {correct_total/len(preds)*100:.0f}%")
print(f"\n  Saved → /content/training_curves.png")
print(f"  Saved → /content/test_results.png")
print("\n✅ Complete pipeline done!")
```

### Visual Test Results for All Sample Images

The following displays the visual test results for all images in the `visual_test` directory, including the true label, predicted label, and confidence. A green border indicates a correct prediction, while a red border indicates an incorrect prediction.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# ── 1. Define Mappings (Adjusted to catch plurals) ───────────────
# This ensures that even if r['pred'] is "cats", it finds the icon
EMOJI = {
    'cat': '🐱',
    'dog': '🐶'
}

# ── 2. Draw grid ─────────────────────────────────────────────────
n_cols = 5
n_rows = (len(preds) + n_cols - 1) // n_cols

fig, axes = plt.subplots(
    n_rows, n_cols,
    figsize=(n_cols * 4, n_rows * 4.5)
)

fig.suptitle(
    'ResNet-18 — Cat vs Dog  |  Visual Test Images\n'
    'Green border = Correct Prediction   |   Red border = Wrong Prediction',
    fontsize=13, fontweight='bold', y=1.01
)

correct_total = 0

for idx in range(n_rows * n_cols):
    row = idx // n_cols
    col = idx % n_cols

    # Handle single row/col cases
    if n_rows > 1:
        ax = axes[row][col]
    else:
        ax = axes[col]

    ax.axis('off')

    if idx >= len(preds):
        continue

    r = preds[idx]
    ax.imshow(r['img'])

    # ── CRITICAL FIX: Normalize strings for comparison ──────────
    # This removes capitalization and trailing 's' (e.g., "Dogs" -> "dog")
    true_norm = str(r['true']).lower().rstrip('s')
    pred_norm = str(r['pred']).lower().rstrip('s')

    is_actually_correct = (true_norm == pred_norm)
    # ────────────────────────────────────────────────────────────

    # Colored border using the normalized check
    border_color = '#2ECC71' if is_actually_correct else '#E74C3C'
    rect = plt.Rectangle((0,0), 1, 1, fill=False, edgecolor=border_color,
                         linewidth=12, transform=ax.transAxes)
    ax.add_patch(rect)

    # Title text
    text_color = '#1A8A35' if is_actually_correct else '#C0392B'
    status = '✅' if is_actually_correct else '❌'

    # Get emoji using normalized name
    true_icon = EMOJI.get(true_norm, r['true'])
    pred_icon = EMOJI.get(pred_norm, r['pred'])

    ax.set_title(
        f"{status}  True : {true_icon}\n"
        f"     Pred : {pred_icon}\n"
        f"     Conf : {r['conf']*100:.1f}%",
        fontsize=10, color=text_color, fontweight='bold', pad=10
    )

    if is_actually_correct:
        correct_total += 1

# Legend
legend_elements = [
    mpatches.Patch(color='#185FA5', label='🐱 Cat'),
    mpatches.Patch(color='#E24B4A', label='🐶 Dog'),
    mpatches.Patch(color='#2ECC71', label='✅ Correct'),
    mpatches.Patch(color='#E74C3C', label='❌ Wrong')
]
fig.legend(handles=legend_elements, loc='lower center', ncol=4, bbox_to_anchor=(0.5, -0.03))

plt.tight_layout()
plt.show()

# ── 3. Summary Table ─────────────────────────────────────────────
print(f"\n{'='*55}\nVISUAL TEST RESULTS SUMMARY\n{'='*55}")
print(f"{'#':<4} {'True':<8} {'Predicted':<10} {'Confidence':<12} {'Result'}")
print("-" * 55)

for i, r in enumerate(preds):
    # Use the same normalized logic here
    t_n = str(r['true']).lower().rstrip('s')
    p_n = str(r['pred']).lower().rstrip('s')
    res_text = "✅ Correct" if t_n == p_n else "❌ Wrong"

    print(f"{i+1:<4} {r['true']:<8} {r['pred']:<10} "
          f"{r['conf']*100:>6.1f}%      {res_text}")

print(f"\n  Final Score: {correct_total} / {len(preds)}")
```
